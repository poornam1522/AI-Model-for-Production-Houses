```python
def get_paraphrase_pairs():
  response = openai.Completion.create(
  model="text-davinci-002",
  prompt="The following is an original sentence followed by a paraphrased version of it with a diverse choice of words:\n\noriginal: Once, a group of frogs was roaming around the forest in search of water.\nparaphrase: A herd of frogs was wandering around the woods in search of water.\n###\noriginal: Puppies were adopted by numerous kind souls at the puppy drive.\nparaphrase: Many kind souls adopted puppies during the puppy drive.\n###\noriginal: Symptoms of influenza include fever and nasal congestion.\nparaphrase: A stuffy nose and elevated temperature are signs you may have the flu.\n###\noriginal:",
  temperature=0.85,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0.61,
  presence_penalty=0.62,
  stop=["###"]
)
  output = response['choices'][0]['text']
  original,paraphrase = output.split("paraphrase:") 
  original = original.strip()
  paraphrase = paraphrase.strip()
  return original,paraphrase
```

```python
import gradio
import gradio as gr
import pandas as pd

paraphrase_file_path = "/content/drive/My Drive/Practical Introduction to NLP Course/Module 4/paraphrases.csv"

def get_paraphrases(text):
    orig,para = get_paraphrase_pairs()
    out = orig + '\n\n'+para
    return out

def save_to_dataframe(text):
    df = pd.read_csv(paraphrase_file_path)
    ori,para =text.split('\n\n')
    row = [ori,para]
    df.loc[len(df)] = row 
    df.to_csv(paraphrase_file_path, index=False)
    
    return "Saved!"
    

with gr.Blocks() as demo:
    textbox = gr.Textbox(placeholder="Press generate button to get paraphrases", lines=5)
    btn1 = gr.Button("Generate Paraphrases")
    btn2 = gr.Button("Save")

    btn1.click(get_paraphrases,inputs=textbox,outputs=textbox)
    btn2.click(save_to_dataframe, inputs=textbox, outputs=textbox)

demo.launch(debug=True)
```
