```python
def getparaphrase(sentence):
  response = openai.Completion.create(
  model="text-davinci-002",
  prompt="The following is an original sentence followed by a paraphrased version of it with a diverse choice of words:\n\noriginal: Once, a group of frogs was roaming around the forest in search of water.\nparaphrase: A herd of frogs was wandering around the woods in search of water.\n###\noriginal: Puppies were adopted by numerous kind souls at the puppy drive.\nparaphrase: Many kind souls adopted puppies during the puppy drive.\n###\noriginal: Symptoms of influenza include fever and nasal congestion.\nparaphrase: A stuffy nose and elevated temperature are signs you may have the flu.\n###\noriginal: {0}\nparaphrase:".format(sentence),
  temperature=0.85,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0.72,
  presence_penalty=0.72,
  stop=["###"]
)
  paraphrased_sentence = response['choices'][0]['text']
  paraphrased_sentence = paraphrased_sentence.strip()
  return paraphrased_sentence
```

```python
# sentence1 = "Elon Musk has shown again he can influence the digital currency market with just his tweets."
sentence2 = "It triggered a downward spiral for Bitcoin value but the cryptocurrency has stabilised since."
paraphrased = getparaphrase(sentence2)
print (paraphrased)
```

# Text Paraphraser

```python
input_text = """Elon Musk has shown again he can influence the digital currency market with just his tweets. 
After saying that his electric vehicle-making company Tesla will not accept payments in Bitcoin because of environmental concerns, he tweeted that he was working with developers of Dogecoin to improve system transaction efficiency. 

Following the two distinct statements from him, the world's largest cryptocurrency hit a two-month low, while Dogecoin rallied by about 20 percent. 
The SpaceX CEO has in recent months often tweeted in support of Dogecoin, but rarely for Bitcoin.  
In a recent tweet, Musk put out a statement from Tesla that it was concerned about the rapidly increasing use of fossil fuels for Bitcoin (price in India) mining and transaction, and hence was suspending vehicle purchases using the cryptocurrency."""
```

```python
!python -m spacy download en_core_web_sm
```

```python
import spacy
nlp = spacy.load("en_core_web_sm",disable = ['ner', 'tagger','lemmatizer','textcat'])
```

```python
sentences = nlp(input_text).sents
```

```python
for sentence in sentences:
  stripped_sentence = sentence.text.strip()
  if stripped_sentence!="":
    print (stripped_sentence)
```

```python
def getsentences(txt):
  sentences = [i.text.strip() for i in nlp(txt).sents if i.text.strip()!=""]
  return sentences
sents = getsentences(input_text)
```

```python
print (sents)
```

```python
from concurrent.futures import ThreadPoolExecutor as Executor
with Executor() as executor:
        futures = [executor.submit(getparaphrase, s) for s in sents]
loop_all_answers = [f.result() for f in futures]
```

```python
for p in loop_all_answers:
  print (p)
```

```python
for orig in sents:
  print (orig)
```

```python
full_paraphrased_text = ' '.join(loop_all_answers)
print (full_paraphrased_text)
```
