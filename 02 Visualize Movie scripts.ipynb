```python
movie_plots_path =  "/content/drive/My Drive/Practical Introduction to NLP Course/Module1/American_Movie_Plots_2005_2021_v1.csv"

import pandas as pd 
df = pd.read_csv(movie_plots_path) 
print (df.shape)
df.head()
```

```python
from sentence_transformers import SentenceTransformer
# all-MiniLM-L6-v2
# all-mpnet-base-v2
# https://www.sbert.net/docs/pretrained_models.html
sentence_encoder_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
print (df.shape)
df.drop_duplicates(subset ="Wiki Link", keep = False, inplace = True)
print (df.shape)
```

```python
movie_plots = df['Plot'].tolist()
print (len(movie_plots))
print (movie_plots[0])
```

```python

embeddings = sentence_encoder_model.encode(movie_plots,show_progress_bar=True)
```

```python
print (len(embeddings[0]))
print (embeddings[0])
```

```python
df['full_embeddings']= embeddings.tolist()
df.head(3)
```

```python
import umap
import numpy as np

# embeddings - np.array
umap_embeddings_2dim = umap.UMAP(n_neighbors=5, 
                            n_components=2, 
                            min_dist=0.0,metric='cosine').fit_transform(embeddings)

df["umap_embeddings_2dim"]  = umap_embeddings_2dim.tolist()
df.head(3)
```

https://umap-learn.readthedocs.io/en/latest/parameters.html

```python
import numpy as np
import plotly.express as px

col="umap_embeddings_2dim"

hover_data =["Wiki Link"]
hover_name = "Movie Name"
plot_values = np.stack(df[col], axis=1)


x, y = plot_values[0], plot_values[1]
fig = px.scatter(
            df,
            x=x,
            y=y,
            hover_data=hover_data,
            title="Visualize Movie Scripts",
            hover_name=hover_name,
        )

fig.show()
```

```python
directors_desc_path = "/content/drive/My Drive/Practical Introduction to NLP Course/Module 2/top_100_directors_description.csv"

import pandas as pd 
df_dirs = pd.read_csv(directors_desc_path) 
print (df_dirs.shape)
df_dirs.head(5)
```

```python
dir_bios = df_dirs['selection1_Biography'].tolist()
print (dir_bios[0])
```

```python
embeddings_dirs = sentence_encoder_model.encode(dir_bios,show_progress_bar=True)
df_dirs['full_embeddings']= embeddings_dirs.tolist()
df_dirs.head(3)
```

```python
df_dirs['type']='Director'
df_dirs.head()
```

```python
df_directors = df_dirs.rename(columns={'selection1_name': 'Name', 'selection1_url': 'Link'})
df_directors =   df_directors[["Name", "Link","full_embeddings","type"]]
df_directors.head(3)
```

```python
df_movies = df.rename(columns={'Movie Name': 'Name', 'Wiki Link': 'Link'})
df_movies =   df_movies[["Name", "Link","full_embeddings","type"]]
df_movies.head(3)
```

```python
combined_df = [df_movies, df_directors]
combined_df_result = pd.concat(combined_df)
combined_df_result.sample(25)
```

```python

full_embeddings_list = np.array(combined_df_result['full_embeddings'].tolist())
```

```python
# embeddings - np.array
umap_embeddings_2dim_combined = umap.UMAP(n_neighbors=5, 
                            n_components=2, 
                            min_dist=0.0,metric='cosine').fit_transform(full_embeddings_list)
```

```python
combined_df_result["umap_embeddings_2dim"]  = umap_embeddings_2dim_combined.tolist()
combined_df_result.head(3)
```

```python
import numpy as np
import plotly.express as px

col="umap_embeddings_2dim"

hover_data =["Link"]
hover_name = "Name"
plot_values = np.stack(combined_df_result[col], axis=1)
color = "type"

x, y = plot_values[0], plot_values[1]
fig = px.scatter(
            combined_df_result,
            x=x,
            y=y,
            color = color,
            hover_data=hover_data,
            title="Visualize Movie Scripts and Directors",
            hover_name=hover_name
        )

fig.show()
```
